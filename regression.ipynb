```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
class LinearRegression():    
    def __init__(self,num_coef = 1):
        self.num_coef = num_coef
        self.coefs = np.ones(num_coef)
        self.intercept = np.ones(1)
    
    def fit(self, X, y, iterations = 100,lr=0.1):
        N = len(X)
        for _ in range(iterations):
            for i in range(len(X)):
                ground_truth = y[i]                
                prediction = np.dot(X[i],self.coefs) + self.intercept
                gradient = (2/N) * (prediction - ground_truth) * X[i] 
                gradient_i = (2/N) * (prediction - ground_truth)
                self.coefs += lr * (-gradient)
                self.intercept += lr * (-gradient_i)
    
    def predict(self, X):
        y = []
        for x in X:
            y.append(np.dot(x,self.coefs) + self.intercept)
        return np.array(y)
    
    def error(self, X, y):
        prediction = self.predict(X)
        return np.sum((y - prediction)**2 / len(y))
    
    def get_params(self):
        return self.coefs, self.intercept
        
```

```python
test = LinearRegression(num_coef=1)
X = np.array([[1],[2],[3],[4],[5]])
y = np.array([4,9,10,11,15])
test.fit(X, y, iterations=1000, lr=0.1)
prediction = test.predict(X)
test.error(X,y)
```

```python
plt.plot(X,y,label='Ground Truth')
plt.plot(X,prediction,label='Predicted')
plt.legend()
plt.show()
```
